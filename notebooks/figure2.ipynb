```python
# 绘制多个热力图到一行中
fig, axes = plt.subplots(1, 4, figsize=(20, 5), gridspec_kw={'width_ratios': [1, 1, 1, 1], 'wspace': 0.3})

# 数据和标题的映射
datasets = [
    (log_df, "Llama-2-7b"),
    (log_updated_df, "Llama-3.1-8B"),
    (log_final_df, "Qwen2-7B"),
    (log_new_df, "Mistral-7B-v0.1")
]

# 遍历绘制每个子图
for ax, (data, title) in zip(axes, datasets):
    heatmap = ax.imshow(data, cmap="coolwarm", interpolation="nearest")
    ax.set_title(title, fontsize=14)
    ax.set_xlabel("Window Size", fontsize=14)
    if title == "Llama-2-7b":
        ax.set_ylabel("Eval Length", fontsize=14)
    ax.set_xticks(range(len(data.columns)))
    ax.set_xticklabels(data.columns, rotation=45)
    ax.set_yticks(range(len(data.index)))
    ax.set_yticklabels(data.index)

# 添加共享的颜色条
cbar = fig.colorbar(heatmap, ax=axes, location="right", shrink=0.8, pad=0.03, label="Log10(Perplexity)")

# 调整布局
plt.tight_layout()
# plt.show()

# 保存到pdf
plt.savefig("four_heatmaps.pdf")
```
